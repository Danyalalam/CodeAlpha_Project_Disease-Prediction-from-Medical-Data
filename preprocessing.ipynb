# Upsampling

```python
# X -> indepenet: input features
# y -> depenedet: output
X = final_df.drop('Outcomes', axis=1)
Y = final_df['Outcomes']
```

```python
X.shape
```

```python
from imblearn.over_sampling import SMOTE

su = SMOTE(random_state=42)
X_su, y_su = su.fit_resample(X, Y)
```

```python
upsample_df = X_su

upsample_df['Outcomes'] = y_su
print(upsample_df.shape)
upsample_df.head()
```

```python
upsample_df.groupby('Outcomes').size().plot(kind='pie',
                                       y = "Outcomes",
                                       label = "Type",
                                       autopct='%1.1f%%')
```

```python
y_su.value_counts()
```
